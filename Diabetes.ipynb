```python
import sklearn
from sklearn.datasets import load_boston
import pandas as pd
bos_hou = load_boston()
 
Q1 = np.percentile(df['BMI'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(df['BMI'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", df.shape)
 
# Upper bound
upper = np.where(df['BMI'] >= (Q3+1.5*IQR))
# Lower bound
lower = np.where(df['BMI'] <= (Q1-1.5*IQR))
 
''' Removing the Outliers '''
df.drop(upper[0], inplace = True)
df.drop(lower[0], inplace = True)
 
print("New Shape: ", df.shape)
```

```python
df = df.drop(['Education', 'Veggies' , 'Fruits' , 'Income'], axis = 1)
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df)
```

```python
X = df.drop(columns='Diabetes_binary', axis=1)
Y = df['Diabetes_binary']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size =.2,random_state=42)
```

```python
X_train =scaler.fit_transform(X_train)
X_test =scaler.fit_transform(X_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler(feature_range=(-1,1)).fit(X_train)
X_train = scaling.transform(X_train)
X_test = scaling.transform(X_test)
```

```python
from sklearn.linear_model import  LogisticRegression
```

```python
log = LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
log.fit(X_train , Y_train)
```

```python
log_predict =log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test , log_predict)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
print(confusion_matrix(Y_test , log_predict))
```

```python
print(classification_report(Y_test , log_predict))
```

```python
from sklearn import svm
```

```python
 svm_model.fit(X_train, Y_train)
```

```python
svm_model_predict = svm_model.predict(X_test)
```

```python
 accuracy_score(Y_test , svm_model_predict)
```

```python
print(classification_report(Y_test , svm_model_predict))
```
